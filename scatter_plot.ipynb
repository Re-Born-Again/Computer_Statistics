```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import csv
```

```python
df=pd.read_csv('mtcars.csv')
for i in list(df.head())[2:]:
 plt.scatter(df['mpg'],df[i])
 plt.show()
```

```python
with open('mtcars.csv','r') as csvfile:
	lines = csv.reader(csvfile, delimiter=',')
	Names=[]
	Values=[]
	for row in lines:
		Names.append(row[0])
		Values.append(row[1])

plt.scatter(Names, Values, color = 'b',edgecolor='r',s = 50)
plt.xticks(rotation = 75)
plt.xlabel('Model')
plt.ylabel('MPG')
plt.title('MTcars', fontsize = 20)

plt.show()
```
